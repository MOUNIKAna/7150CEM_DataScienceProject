```python
df = df.astype({'YEAR OCC': 'int64', 'MONTH OCC': 'int64', 'DAY OCC': 'int64'})
```

```python
numeric_columns = df.select_dtypes(include=['int64', 'float64'])
plt.figure(figsize=(10, 8))
plt.title('Correlation')
sns.heatmap(numeric_columns.corr(), annot=True, linewidths=.5, cmap="YlGnBu")
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, LabelEncoder, OneHotEncoder
df_scaled = df.copy()
excluded_features = ['MONTH OCC', 'AREA NAME', 'Vict Sex', 'Vict Descent', 'Status'] # features on this list will not be scaled
selected_features = np.array(df.drop(excluded_features, axis=1, inplace=False).columns).reshape(-1, 1)

for feature in selected_features:
    scaler = MinMaxScaler()
    scaler.fit(df_scaled[feature])
    df_scaled[feature] = scaler.transform(df_scaled[feature])
```

```python
df_train = df.drop(['AREA NAME', 'Vict Descent', 'Status'] ,axis=1, inplace=False)
df_train = pd.get_dummies(df_train)
```

```python
df_train
```

```python
features = np.array(df_train.columns).reshape(-1, 1)

for feature in features:
    scaler = MinMaxScaler()
    scaler.fit(df_train[feature])
    df_train[feature] = scaler.transform(df_train[feature])
```

```python
train_features = df[['YEAR OCC', 'MONTH OCC', 'DAY OCC', 'TIME OCC', 'AREA', 'Vict Age', 'Premis Cd Log', 'Weapon Used Cd Log']]
train_label = df['Crm Cd Log'].astype(int)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_features, train_label, test_size=0.2, random_state=11)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, roc_auc_score
from sklearn.linear_model import LogisticRegression
```

```python
lr_reg = LogisticRegression(solver='liblinear')
lr_reg.fit(X_train, y_train)
```

```python
lr_preds = lr_reg.predict(X_test)
lr_preds_proba = lr_reg.predict_proba(X_test)
lr_mse = mean_squared_error(y_test, lr_preds)
lr_rmse = np.sqrt(lr_mse)
lr_r2 = abs(r2_score(y_test, lr_preds))
lr_acc = accuracy_score(y_test, lr_preds)
lr_roc_auc = roc_auc_score(y_test, lr_preds_proba, multi_class='ovr')
```

```python
print('Measuring the performance : MSE : {0:.3f}'.format(lr_mse))
print('Measuring the performance : RMSE : {0:.3f}'.format(lr_rmse))
print('Variance Score : {0:.3f}'.format(lr_r2))
print('Accuracy Score : {0:.3f}'.format(lr_acc))
print('ROC_AUC Score : {0:.3f}'.format(lr_roc_auc))
```

```python
lr_classification_report = classification_report(y_test, lr_preds)
print("Classification Report for Logistic Regression:\n", lr_classification_report)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
```

```python
import xgboost as xgb
```

```python
class_mapping = {4: 0, 5: 1, 6: 2}
y_train_mapped = [class_mapping[label] for label in y_train]
y_test_mapped = [class_mapping[label] for label in y_test]
```

```python
# To improve the speed of model performance
xgb_model = xgb.XGBClassifier(n_estimators=1000, learning_rate=0.01)
xgb_model.fit(X_train, y_train_mapped)
```

```python
xgb_preds = xgb_model.predict(X_test)
xgb_preds_proba = xgb_model.predict_proba(X_test)
```

```python
xgb_mse = mean_squared_error(y_test_mapped, xgb_preds)
xgb_rmse = np.sqrt(xgb_mse)
xgb_r2 = r2_score(y_test_mapped, xgb_preds)
xgb_acc = accuracy_score(y_test_mapped, xgb_preds)
xgb_roc_auc = roc_auc_score(y_test_mapped, xgb_preds_proba, multi_class='ovr')
```

```python
print('MSE : {0:.3f}'.format(xgb_mse))
print('RMSE : {0:.3f}'.format(xgb_rmse))
print('Variance Score : {0:.3f}'.format(xgb_r2))
print('Accuracy Score : {0:.3f}'.format(xgb_acc))
print('ROC_AUC Score : {0:.3f}'.format(xgb_roc_auc))
```

```python
xgb_classification_report = classification_report(y_test_mapped, xgb_preds)
print("Classification Report for XGBoost:\n", xgb_classification_report)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_model = GradientBoostingClassifier()
gb_model.fit(X_train, y_train)
```

```python
gb_preds = gb_model.predict(X_test)
gb_preds_proba = gb_model.predict_proba(X_test) 
```

```python
gb_mse = mean_squared_error(y_test, gb_preds)
gb_rmse = np.sqrt(gb_mse)
gb_r2 = r2_score(y_test, gb_preds)
gb_acc = accuracy_score(y_test, gb_preds)
gb_roc_auc = roc_auc_score(y_test, gb_preds_proba, multi_class='ovr')
```

```python
print('MSE : {0:.3f}'.format(gb_mse))
print('RMSE : {0:.3f}'.format(gb_rmse))
print('Variance Score : {0:.3f}'.format(gb_r2))
print('Accuracy Score : {0:.3f}'.format(gb_acc))
print('ROC_AUC Score : {0:.3f}'.format(gb_roc_auc))
```

```python
gb_classification_report = classification_report(y_test, gb_preds)
print("Classification Report for GradientBoosting:\n", gb_classification_report)
```

```python
results = {
    'Model': ['Logistic Regression', 'XGBClassifier', 'GBCLassifier'],
    'MSE': [lr_mse, xgb_mse, gb_mse],
    'RMSE': [lr_rmse, xgb_rmse, gb_rmse],
    'R^2': [lr_r2, xgb_r2, gb_r2],
    'Accuracy': [lr_acc, xgb_acc, gb_acc],
    'ROC AUC': [lr_roc_auc, xgb_roc_auc, gb_roc_auc]
}

results_df = pd.DataFrame(results)
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
melted_df = pd.melt(results_df, id_vars=['Model'], var_name='Metric', value_name='Value')
sns.barplot(x='Model', y='Value', hue='Metric', data=melted_df, palette="Set3")
plt.title('Comparison of Metrics for Multiple Models')
plt.ylabel('Value')
plt.legend(title='Metrics', loc='upper left')

ax = plt.gca()
for p in ax.patches:
    ax.annotate(f'{p.get_height():.2f}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=10, color='black', xytext=(0, 5),
                textcoords='offset points')

plt.show()
```
